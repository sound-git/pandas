Concatenate dataframes

```python
users = pd.DataFrame({'userid':[1,2,3],'name':['abhi','satish','deepti']})
```

```python
msgs = pd.DataFrame({'userid':[1,2,3,4],'msg':['hello','hi','bye0','hey']})
```

```python
pd.concat([users,msgs])
#if you notice, original index of rows in each dataframe is preserved
```

```python
#what if we need the index to follow the sequence?
pd.concat([users,msgs],ignore_index=True)
```

```python
#concat horizantally instead of vertical
pd.concat([users,msgs],axis=1)
#datatype of userid col changed to float. why ? Datatype of NaN is float, hence entire col got changed to float
```

```python
import numpy as np
type(np.nan)
```

```python
#Merging the dataframe
#whats the diff between concatenating vs merging
#concat -> just concat either in axis =0 or 1
#merge -> join based on common cols
  # 1. Inner Join
  # 2. Left Join
  # 3. Right Join
  # 4. Full Outer Join

users.merge(msgs, on='userid')  # here by default inner join applied
```

```python
users.merge(msgs, on='userid', how='left')
```

```python
users.merge(msgs, on='userid', how='outer')
```

```python
users.merge(msgs, on='userid', how='right')
```

```python
#if col names are different
#lets rename to simulate
users.rename(columns={'userid':'id'},inplace=True)
```

```python
users.merge(msgs, left_on='id', right_on='userid', how='right')
```
